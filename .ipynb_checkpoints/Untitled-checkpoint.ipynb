FOOD PRICE PREDICTION PROJECT

Importing all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error
from pmdarima import auto_arima 
```

Loading dataset 

```python
file_path = r"C:\Users\JUDE PRO\Downloads\archive\wfp_food_prices_uga.csv"
df = pd.read_csv(file_path)

# Preview columns
print(df.columns)
df.head()
```

DATA CLEANING
 Convert date column to datetime
- Convert price to numeric
- Remove invalid values

```python
# Clean date
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])

# Clean price
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])

df.head()
```

DATASET FILTERING
Keep data from 2017 to 2021
- Select important columns only
- Focus on selected commodities

```python
# Filter years
df = df[
    (df['date'].dt.year >= 2017) &
    (df['date'].dt.year <= 2021)
]

# Keep important columns
df = df[['date', 'commodity', 'price']]

# Filter commodities
selected_commodities = ['Beans', 'Sorghum', 'Maize (white)']
df = df[df['commodity'].isin(selected_commodities)]

df.head()
```

DATA AGGREGATE AND HANDLING MISSING VALUES

```python
# Sort
df = df.sort_values(by='date')

# Group by date and commodity
df = df.groupby(['date', 'commodity'])['price'].mean().reset_index()

df.head()


fixed_data = []

for commodity in selected_commodities:
    temp = df[df['commodity'] == commodity].copy()
    temp = temp.set_index('date')

    full_dates = pd.date_range(start='2017-01-15', end='2021-12-15', freq='MS') + pd.DateOffset(days=14)

    temp = temp.reindex(full_dates)
    temp['commodity'] = commodity

    # Fill missing values
    temp['price'] = temp['price'].interpolate(method='linear')

    temp = temp.reset_index().rename(columns={'index': 'date'})
    fixed_data.append(temp)

# Combine
df = pd.concat(fixed_data)

# Sort
df = df.sort_values(['commodity', 'date'])

df.head()
```

PREPARATION OF DATA FOR SARIMA 
Split dataset into individual time series for each commodity and saving it for further modelling.

```python
# Beans
beans_df = df[df['commodity'] == 'Beans'].copy()
beans_df.set_index('date', inplace=True)
beans_df = beans_df[['price']]

# Maize
maize_df = df[df['commodity'] == 'Maize (white)'].copy()
maize_df.set_index('date', inplace=True)
maize_df = maize_df[['price']]

# Sorghum
sorghum_df = df[df['commodity'] == 'Sorghum'].copy()
sorghum_df.set_index('date', inplace=True)
sorghum_df = sorghum_df[['price']]

print("Beans:")
print(beans_df.head())

print("\nMaize:")
print(maize_df.head())

print("\nSorghum:")
print(sorghum_df.head())

beans_df.to_csv(r"C:\Users\JUDE PRO\Downloads\archive\beans_time_series.csv")
maize_df.to_csv(r"C:\Users\JUDE PRO\Downloads\archive\maize_time_series.csv")
sorghum_df.to_csv(r"C:\Users\JUDE PRO\Downloads\archive\sorghum_time_series.csv")

print("✅ All datasets saved successfully!")
```

SARIMA MODELLING BEANS,SORGHUM AND MAIZE(white)

differencing and ACF and PACF plots

```python
commodities = ['Beans', 'Maize (white)', 'Sorghum']

for commodity in commodities:

    print("\n" + "="*60)
    print(f"📊 {commodity} - ACF / PACF")
    print("="*60)

    data = df[df['commodity'] == commodity].copy()
    data = data.set_index('date').sort_index()

    ts = data['price'].interpolate().ffill().bfill()

    # Differencing
    ts_diff = ts.diff().dropna()

    # ==============================
    # ACF / PACF PLOTS
    # ==============================
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plot_acf(ts_diff, ax=plt.gca(), lags=20)
    plt.title(f"{commodity} - ACF")

    plt.subplot(1,2,2)
    plot_pacf(ts_diff, ax=plt.gca(), lags=20)
    plt.title(f"{commodity} - PACF")

    plt.tight_layout()
    plt.show()
```

TRAIN + AUTO ARIMA (ALL COMMODITIES)

```python
models = {}
splits = {}

for commodity in commodities:
    print("\n" + "="*60)
    print(f"📊 TRAINING: {commodity}")
    print("="*60)

    # 1. Filter and setup index
    data = df[df['commodity'] == commodity].copy()
    data['date'] = pd.to_datetime(data['date'])
    data = data.set_index('date').sort_index()

    # 2. Ensure time series format
    data = data.asfreq('D')

    # safer filling
    ts = data['price'].ffill().bfill().dropna()

    # 3. Train/Test split
    split = int(len(ts) * 0.8)
    train = ts.iloc[:split]
    test = ts.iloc[split:]

    splits[commodity] = (train, test)

    # 4. Auto ARIMA (SAFE MODE → avoids warnings)
    print("🔍 Running Auto ARIMA...")

    auto_model = auto_arima(
        train,
        seasonal=False,      # ✅ IMPORTANT FIX (removes divide-by-zero warnings)
        stepwise=True,
        suppress_warnings=True,
        error_action="ignore",
        max_p=3,
        max_q=3,
        max_d=2
    )

    print(f"Best Order: {auto_model.order}")

    # 5. Fit ARIMA (NOT SARIMAX → avoids instability)
    model = SARIMAX(
        train,
        order=auto_model.order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    # 6. Fit model safely
    try:
        results = model.fit(
            disp=False,
            maxiter=500,
            method='lbfgs'
        )

        models[commodity] = results

    except Exception as e:
        print(f"❌ Model failed for {commodity}: {e}")
```

FORECAST AND EVALUATION

```python
results_summary = []

for commodity in commodities:

    print("\n" + "="*60)
    print(f"📊 FORECAST: {commodity}")
    print("="*60)

    train, test = splits[commodity]
    results = models[commodity]

    # Forecast
    forecast_values = results.forecast(steps=len(test))
    forecast = pd.Series(forecast_values, index=test.index)

    # Safe alignment
    combined = pd.concat([test, forecast], axis=1)
    combined.columns = ["actual", "forecast"]
    combined = combined.dropna()

    if len(combined) == 0:
        print("❌ No valid data:", commodity)
        continue

    test_clean = combined["actual"]
    forecast_clean = combined["forecast"]

    # Metrics
    mae = mean_absolute_error(test_clean, forecast_clean)

    mape = np.mean(
        np.abs((test_clean - forecast_clean) / np.maximum(test_clean, 1e-5))
    ) * 100

    print(f"✅ MAE: {mae:.2f}")
    print(f"✅ MAPE: {mape:.2f}%")

    results_summary.append([commodity, mae, mape])

    # ==============================
    # PLOT FORECAST
    # ==============================
    plt.figure(figsize=(12,5))
    plt.plot(train, label="Train")
    plt.plot(test_clean, label="Actual")
    plt.plot(forecast_clean, label="Forecast", linestyle="--")
    plt.title(f"{commodity} SARIMA Forecast")
    plt.legend()
    plt.show()

    # ==============================
    # RESIDUALS
    # ==============================
    residuals = test_clean - forecast_clean

    plt.figure(figsize=(10,4))
    plt.plot(residuals)
    plt.axhline(0, linestyle="--")
    plt.title(f"{commodity} Residuals")
    plt.show()

# FINAL TABLE
results_df = pd.DataFrame(results_summary, columns=["Commodity", "MAE", "MAPE"])
print("\n📊 FINAL RESULTS:")
print(results_df)
```
