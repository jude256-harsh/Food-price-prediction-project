TRAIN + AUTO ARIMA (ALL COMMODITIES)

```python
models = {}
splits = {}

for commodity in commodities:
    print("\n" + "="*60)
    print(f"📊 TRAINING: {commodity}")
    print("="*60)

    # 1. Filter and setup index
    data = df[df['commodity'] == commodity].copy()
    data['date'] = pd.to_datetime(data['date'])
    data = data.set_index('date').sort_index()

    # 2. Ensure time series format
    data = data.asfreq('D')

    # safer filling
    ts = data['price'].ffill().bfill().dropna()

    # 3. Train/Test split
    split = int(len(ts) * 0.8)
    train = ts.iloc[:split]
    test = ts.iloc[split:]

    splits[commodity] = (train, test)

    # 4. Auto ARIMA (SAFE MODE → avoids warnings)
    print("🔍 Running Auto ARIMA...")

    auto_model = auto_arima(
        train,
        seasonal=False,      # ✅ IMPORTANT FIX (removes divide-by-zero warnings)
        stepwise=True,
        suppress_warnings=True,
        error_action="ignore",
        max_p=3,
        max_q=3,
        max_d=2,
    
    )

    print(f"Best Order: {auto_model.order}")

    # 5. Fit ARIMA (NOT SARIMAX → avoids instability)
    model = SARIMAX(
        train,
        order=auto_model.order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    # 6. Fit model safely
    try:
        results = model.fit(
            disp=False,
            maxiter=500,
            method='lbfgs'
        )

        models[commodity] = results

    except Exception as e:
        print(f"❌ Model failed for {commodity}: {e}")
```

FORECAST AND EVALUATION

```python
results_summary = []

for commodity in commodities:

    print("\n" + "="*60)
    print(f"📊 FORECAST: {commodity}")
    print("="*60)

    train, test = splits[commodity]
    results = models[commodity]

    # Forecast
    forecast_values = results.forecast(steps=len(test))
    forecast = pd.Series(forecast_values, index=test.index)

    # Safe alignment
    combined = pd.concat([test, forecast], axis=1)
    combined.columns = ["actual", "forecast"]
    combined = combined.dropna()

    if len(combined) == 0:
        print("❌ No valid data:", commodity)
        continue

    test_clean = combined["actual"]
    forecast_clean = combined["forecast"]

    # Metrics
    mae = mean_absolute_error(test_clean, forecast_clean)

    mape = np.mean(
        np.abs((test_clean - forecast_clean) / np.maximum(test_clean, 1e-5))
    ) * 100

    print(f"✅ MAE: {mae:.2f}")
    print(f"✅ MAPE: {mape:.2f}%")

    results_summary.append([commodity, mae, mape])

    # ==============================
    # PLOT FORECAST
    # ==============================
    plt.figure(figsize=(12,5))
    plt.plot(train, label="Train")
    plt.plot(test_clean, label="Actual")
    plt.plot(forecast_clean, label="Forecast", linestyle="--")
    plt.title(f"{commodity} SARIMA Forecast")
    plt.legend()
    plt.show()

    # ==============================
    # RESIDUALS
    # ==============================
    residuals = test_clean - forecast_clean

    plt.figure(figsize=(10,4))
    plt.plot(residuals)
    plt.axhline(0, linestyle="--")
    plt.title(f"{commodity} Residuals")
    plt.show()

# FINAL TABLE
results_df = pd.DataFrame(results_summary, columns=["Commodity", "MAE", "MAPE"])
print("\n📊 FINAL RESULTS:")
print(results_df)
```

SARIMAX

LOAD AND PREPARATION OF  DATASET 

```python
# Load dataset
df = pd.read_csv("clean_food_price.csv")

# Fix date
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
df.set_index('date', inplace=True)

# Inject fuel prices
fuel_prices = {
    2017: 3549, 
    2018: 4023, 
    2019: 4085, 
    2020: 3851, 
    2021: 4051
}

df['fuel_price'] = df.index.year.map(fuel_prices)

# Clean data
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['fuel_price'] = pd.to_numeric(df['fuel_price'], errors='coerce')

df['price'] = df['price'].interpolate()
df['fuel_price'] = df['fuel_price'].ffill().bfill()

df = df.dropna()

# 🔥 FORCE OUTPUT
print("✅ Data Ready Successfully")
print("\n📊 Data Shape:", df.shape)

print("\n🔍 First 5 Rows:")
print(df[['commodity', 'price', 'fuel_price']].head())

print("\n🔍 Last 5 Rows:")
print(df[['commodity', 'price', 'fuel_price']].tail())
```

DEFINE COMMODITIES

```python
commodities = ['beans', 'sorghum', 'maize (white)']
```

TRAIN-TEST SPLIT

```python
splits = {}

for item in commodities:
    print(f"\n--- {item.upper()} ---")

    item_df = df[df['commodity'].str.strip().str.lower() == item].copy()

    # Monthly aggregation
    item_df = item_df.resample('MS').mean(numeric_only=True)

    # Fill missing
    item_df['price'] = item_df['price'].interpolate()
    item_df['fuel_price'] = item_df['fuel_price'].ffill()

    item_df = item_df.dropna()

    # Split
    split = int(len(item_df) * 0.8)
    train = item_df.iloc[:split]
    test = item_df.iloc[split:]

    print(f"Train shape: {train.shape}, Test shape: {test.shape}")

    splits[item] = (train, test)
```

TRAIN SARIMAX MODELS

```python
models = {}

for item in commodities:
    print(f"\n🔧 TRAINING: {item.upper()}")

    train, test = splits[item]

    # Scaling
    scaler = StandardScaler()
    train_scaled = scaler.fit_transform(train[['price', 'fuel_price']])

    y_train = train_scaled[:, 0]
    exog_train = train_scaled[:, 1].reshape(-1, 1)

    # Auto ARIMA
    auto_model = auto_arima(
        y_train,
        exogenous=exog_train,
        seasonal=False,
        stepwise=True,
        suppress_warnings=True
    )

    print("Best order:", auto_model.order)

    # SARIMAX
    model = SARIMAX(
        y_train,
        exog=exog_train,
        order=auto_model.order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    model_fit = model.fit(disp=False)

    models[item] = (model_fit, scaler)
```

FORECAST AND EVALUATE

```python
for item in commodities:
    print(f"\n📊 FORECAST: {item.upper()}")

    train, test = splits[item]
    model_fit, scaler = models[item]

    # Scale test
    test_scaled = scaler.transform(test[['price', 'fuel_price']])

    y_test = test_scaled[:, 0]
    exog_test = test_scaled[:, 1].reshape(-1, 1)

    # Forecast
    forecast_scaled = model_fit.forecast(
        steps=len(y_test),
        exog=exog_test
    )

    # Inverse scaling
    combined = np.column_stack([forecast_scaled, exog_test.flatten()])
    forecast_real = scaler.inverse_transform(combined)[:, 0]

    actual_combined = np.column_stack([y_test, exog_test.flatten()])
    actual_real = scaler.inverse_transform(actual_combined)[:, 0]

    # Metrics
    mae = mean_absolute_error(actual_real, forecast_real)
    mape = mean_absolute_percentage_error(actual_real, forecast_real)

    print(f"MAE: {mae:.2f}")
    print(f"MAPE: {mape*100:.2f}%")

    # Plot
    plt.figure(figsize=(10,4))
    plt.plot(train.index, train['price'], label="Train")
    plt.plot(test.index, actual_real, label="Actual")
    plt.plot(test.index, forecast_real, label="Forecast", linestyle='--')

    plt.title(f"{item.upper()} SARIMAX Forecast")
    plt.legend()
    plt.grid()
    plt.show()
```
